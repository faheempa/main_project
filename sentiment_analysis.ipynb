```python
from ntscraper import Nitter

scraper = Nitter(log_level=1, skip_instance_check=False)
```

```python
if __name__ == "__main__":
    terms = ["bitcoin", "cryto", "btc", "cryptocurrency"]
    tweets = scraper.get_tweets(terms, mode="term", number=100)
```

```python
from pprint import pprint


pprint(tweets)
```

```python
tweets_list = []
for x in tweets:
    for y in x["tweets"]:
        text = y["text"]
        if text != '':
            tweets_list.append(text)
```

```python

for t in tweets_list:
    print(t)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

def categorize_sentiment(score):
    if score > 0.5:
        return (1, "positive")
    elif score < -0.5:
        return (-1, "negative")
    else:
        return (0, "neutral")

def analyze_sentiment_vader(text):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(text)
    sentiment = categorize_sentiment(score["compound"])
    return sentiment 
```

```python
# test
text = "I love the weather today"
print(analyze_sentiment_vader(text))
```

```python
# test 2
text = "I hate the weather today"
print(analyze_sentiment_vader(text))
```

```python
# test 3
text = "I have no opinion about the weather today"
print(analyze_sentiment_vader(text))
```

```python
result = []
for t in tweets_list:
    result.append(analyze_sentiment_vader(t)[0])

print(result)
```

```python
# get the most frequent sentiment
from collections import Counter

def most_frequent(List):
    occurence_count = Counter(List)
    freq = occurence_count.most_common(1)[0][0]
    if freq == 1:
        return "positive"
    elif freq == -1:
        return "negative"
    else:
        return "neutral"

print(most_frequent(result))
```
